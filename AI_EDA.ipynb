# Exploratory Data Analysis(EDA)

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/drive/MyDrive/AI_Lab_EDA/SalaryPrediction.csv")
df.head(5)
```

```python
# Categorical data
# countplot
sns.countplot(data=df, x='Age')
df['Club'].value_counts()
```

```python

 # pie chart

df['Position'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python

# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['Age'])
```

```python

# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['Age'])
```

```python

# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['Age'])
```

```python

df.describe()
```

```python
# Scatter plot(Numerical - Numerical) for the relation between two column
plt.figure(figsize=(17, 7))
sns.scatterplot(x='Age', y='Apps',hue='Position', style='Club',size='Nation' ,data=df)
plt.show()
```

```python

 # Bar plot (Numerical - Categorical)
  # average age in pclass
plt.figure(figsize=(17, 7))
sns.barplot(x='Age',y='Apps',hue='Position',data=df)
```

```python

#Box plot (Numerical - Categorical)
plt.figure(figsize=(10, 7))
sns.boxplot(x='Position',y='Apps',hue='Age',data=df)
```

```python
# HeatMap (Categorical - Categorical)
sns.heatmap(pd.crosstab(df['Age'],df['Position']))
```

```python


# pairplot
plt.figure(figsize=(10, 7))
sns.pairplot(df,hue='Position')
```

```python
plt.figure(figsize=(10, 7))
sns.lineplot(x='Age',y='Position',data=df)
```
